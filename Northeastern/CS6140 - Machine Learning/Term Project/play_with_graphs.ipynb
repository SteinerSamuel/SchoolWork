```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

csvs = [("run-test2-tag-MSE_train.csv", "Full batch"),
        ("run-test3-tag-MSE_train.csv", "Daily Minibatches")]
```

```python

for file_name, run_name in csvs:
    p = pd.read_csv(file_name)
    p["Wall time"] -= p["Wall time"].min()
    p["Wall time"]/=3600
    plt.plot(p["Wall time"], np.sqrt(p["Value"]))

plt.xlabel("Hours of training")
plt.ylabel("RMSE on training set")
plt.title("Training curves for 14-day dataset (Training)")
plt.legend(list(n for _, n in csvs))

plt.savefig("minibatch_instability_train.png", dpi=300)
```
